```python
#Loading the movie data from CSV file
# Print data from CSV file
df_bom_csv= pd.read_csv("data/bom.movie_gross.csv.gz")
print(df_bom_csv)
```

```python
#Getting the basic stats from the dataset to help with the analysis
print(df_bom_csv.describe)
```

```python
#beginning to cleaning data by checking for missing values
print(df_bom_csv.isnull().sum())
```

```python
#Visualize the distribution of movie gross revenue
plt.figure(figsize=(8, 6))
plt.hist(df_bom_csv["domestic_gross"].dropna(),bins=30,color="blue")
plt.xlabel("Domestic gross revenue(in billions)")
plt.ylabel("frequency")
plt.title("Gross revenue of Movie distribution")
plt.show()
```

```python
# Visualize the top 10 studios with highest average movie gross revenue

top_studios = df_bom_csv.groupby("studio")["domestic_gross"].mean().nlargest(10)
plt.figure(figsize=(10,6))
top_studios.plot(kind="bar", color="orange")
plt.xlabel("studio")
plt.ylabel('Average Domestic Gross Revenue (in billions)')
plt.title('Top 10 Studios with Highest Average Movie Gross Revenue')
plt.show()
```

```python
# Perform data analysis to identify genres with high demand
genre_demand = df_imdb['genre'].value_counts().sort_values(ascending=False)
```

```python
# Get genres with highest demand
top_genres = genre_demand.head(3)  # Replace 3 with desired number of top genres
```

```python
# Filter movies based on top genres
df_top_movies = df_imdb[df_imdb['genre'].isin(top_genres.index)]
```

```python
# Print top movies in genres with high demand
print("Top movies in genres with high demand:")
print(df_top_movies)
```
